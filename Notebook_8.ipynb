```python
import numpy as np
# Importing standard Qiskit libraries
from qiskit import QuantumCircuit, execute, transpile, Aer, IBMQ
from qiskit.tools.jupyter import *
from qiskit.visualization import *
from ibm_quantum_widgets import *

# Loading your IBM Quantum account(s)
provider = IBMQ.load_account()
print('Libraries imported successfully!')
```

```python
# Quantum circuit with 2 qubits
qc = QuantumCircuit(2)
```

```python
# Adding X gate to qubit 0.
qc.x(0)
qc.draw()
```

```python
# Simulating using Statevector simulator 
backend = Aer.get_backend('statevector_simulator') 
job = execute(qc,backend) 
result = job.result()

counts = result.get_counts()
plot_histogram(counts)
```

```python
# Quantum circuit with 2 qubits. Adding an H gate to qubit 0 and an X gate to qubit 1

qc = QuantumCircuit(2)
qc.h(0)
qc.x(1)
qc.draw()
```

```python
# Simulating using Statevector simulator 
backend = Aer.get_backend('statevector_simulator') 
job = execute(qc,backend) 
result = job.result()

counts = result.get_counts()
plot_histogram(counts)
```

```python
# Quantum circuit with 2 qubits. 
# Adding X gates to both qubits, and then a CX gate with qubit 0 as control and qubit 1 as target
qc = QuantumCircuit(2)
qc.x(0)
qc.x(1)
qc.cx(0,1)
qc.draw()
```

```python
# Simulating using Statevector simulator 
backend = Aer.get_backend('statevector_simulator') 
job = execute(qc,backend) 
result = job.result()

counts = result.get_counts()
plot_histogram(counts)
```

```python
# Simulating using Statevector simulator 
backend = Aer.get_backend('statevector_simulator') 
job = execute(qc,backend) 
result = job.result()

state = result.get_statevector()
array_to_latex(state, prefix="\\text{Statevector} = ") # Visualizing as matrix 
```

```python
# 2 Qubit quantum circuit. Adding an X gate to qubit 0
# Simulating using Statevector simulator 
qc = QuantumCircuit(2)
qc.x(0)

backend = Aer.get_backend('statevector_simulator') 
job = execute(qc,backend) 
result = job.result()

counts = result.get_counts()
plot_histogram(counts)
```

```python
# Adding three CX gates to the circuit
# The first CX gate has qubit 0 as control and qubit 1 as target
# The second CX gate has qubit 1 as control and qubit 0 as target
# The third CX gate has qubit 0 as control and qubit 1 as target

qc.cx(0,1)
qc.cx(1,0)
qc.cx(0,1)
qc.draw()
```

```python
# Simulating using Statevector simulator 

backend = Aer.get_backend('statevector_simulator') 
job = execute(qc,backend) 
result = job.result()

counts = result.get_counts()
plot_histogram(counts)
```
